```python
import paramiko
import yaml
from yaml.loader import SafeLoader

# Usuario de la instancia EC2
USUARIO = "ec2-user"

# Llave privada para SSH
LLAVE = paramiko.RSAKey(filename="gbconnected.pem")

servidores = []

# Lista de servidores

with open(("plantas.yaml"), 'r') as yaml_file:
    yaml_content = yaml.load(yaml_file, Loader=SafeLoader)
for ip in yaml_content.values():
    servidores += ip       
    
ssh = paramiko.SSHClient()
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())

for servidor in servidores:
    print(f"Obteniendo 'docker ps' de {servidor[1]}...")
    try:
        ssh.connect(hostname=servidor[0], username=USUARIO, pkey=LLAVE)
        stdin, stdout, stderr = ssh.exec_command("docker ps")
        output = stdout.read().decode()
        with open(f"Shell/{servidor[1]}.txt", "w") as file:
            file.write(output)
        ssh.close()
    except:
        ssh.close()
```
